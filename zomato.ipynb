```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

zomato_data=pd.read_csv("/content/drive/MyDrive/data science/zomato.csv",encoding='latin-1')
```

```python
data_india=zomato_data[zomato_data['Country Code']==1]
```

```python
data_india.info()
```

```python
data_india['City'].value_counts()
```

```python
data=data_india[(data_india['City']=='New Delhi') | (data_india['City']=='Gurgaon') | (data_india['City']=='Noida') | (data_india['City']=='Faridabad')]
```

```python
data.info()
```

data['Restaurant Name'].value_counts()

```python
data.drop(['Country Code','Currency'],axis=1,inplace=True)
```

```python
data.info()
```

```python
for x,y in zip(data['Restaurant Name'].value_counts().index,data['Restaurant Name'].value_counts().values):
  if y<5:
    data.drop(data[data['Restaurant Name']==x].index,axis=0,inplace=True)
```

```python
data['Restaurant Name'].value_counts()
```

```python
plt.figure(figsize=(20,10))
plt.bar(data['Restaurant Name'].value_counts().index,data['Restaurant Name'].value_counts().values)
plt.xticks(rotation=-90)
plt.yticks(np.arange(0,85,2))
plt.ylabel('frequency')
plt.xlabel('Restaurant')
plt.grid(color='r')
plt.show()
```

```python
plt.figure(figsize=(30,6))
plt.bar(data[data['City']=='New Delhi']['Locality'].value_counts().index,data[data['City']=='New Delhi']['Locality'].value_counts().values)
plt.xticks(rotation=-90)
plt.yticks(np.arange(0,25,2))
plt.ylabel('frequency')
plt.xlabel('Locality')
plt.grid(color='y')
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.bar(data[data['City']=='Gurgaon']['Locality'].value_counts().index,data[data['City']=='Gurgaon']['Locality'].value_counts().values)
plt.xticks(rotation=-90)
plt.ylabel('frequency')
plt.xlabel('Locality')
plt.grid(color='g')
plt.show()
```
