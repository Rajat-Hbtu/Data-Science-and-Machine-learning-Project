```python
data.drop(['Last Update'],axis=1,inplace=True)
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder

lbl=LabelEncoder()
data['country']=lbl.fit_transform(data['country'])
```

```python
data['country']
```

```python
data.drop('state',axis=1,inplace=True)
```

```python
data
```

```python
def to_integer(x):
  
  return 1000*x.year + 100*x.month + x.day

# apply the above method on the dataframe
data['date'] = data['date'].apply(lambda x : to_integer(x))
```

```python
data.corr()
```

```python
x=data.drop('Deaths',axis=1)
y=data['Deaths']
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler

s=StandardScaler()

for column in x.columns:
  x[column]=s.fit_transform(x[column].values.reshape(-1,1))
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=20)
```

```python
x_train, y_train
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor

rfr=RandomForestRegressor(verbose=1,n_jobs=-1)

rfr.fit(x_train,y_train)
```

```python
rfr.score(x_test,y_test)
```

```python
predict=rfr.predict(x_test)
```

```python
df=pd.DataFrame({'Actual':y_test,'Prediction':predict,'error':abs(y_test-predict)/y_test*100})
```

```python
from sklearn.neighbors import KNeighborsRegressor

knnr=KNeighborsRegressor()
n_neighbors = [2,3,4,5,6,7]
weights = ['uniform', 'distance']
p = [1,2]

parameters = {'n_neighbors':n_neighbors,
              'weights':weights,
              'p':p}
rscv_knn=RandomizedSearchCV(knnr,param_distributions=parameters,n_iter=100,n_jobs=-1,cv=5,verbose=1,random_state=2)
rscv_knn.fit(x_train,y_train)
```

```python
rscv_knn.score(x_test,y_test)
```
